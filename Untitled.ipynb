# CNN Model

```python
n_features, n_outputs = 50, 7
CNN = models.Sequential()
CNN.add(layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu',
                      input_shape=(n_features,n_features , 1), padding='same'))
CNN.add(layers.MaxPooling2D(pool_size=(2 , 2)))
CNN.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
CNN.add(layers.Dropout(0.5))
CNN.add(layers.Conv2D(filters=128, kernel_size=(5,5), activation='relu'))
CNN.add(layers.MaxPooling2D(pool_size=(2,2)))
CNN.add(layers.Dropout(0.6))
CNN.add(layers.Flatten())
CNN.add(layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.0001)))
CNN.add(layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.0001)))
CNN.add(layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.0001)))
CNN.add(layers.Dense(n_outputs, activation='softmax', kernel_regularizer=regularizers.l2(0.0001)))
```

```python
CNN.summary()
```

```python
CNN.compile(optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy'])
```

```python
callback = [keras.callbacks.EarlyStopping(monitor='accuracy', mode='max', patience=5)]
```

```python
images_df = np.reshape(images_df, (images_df.shape[0], 50, 50, 1))
```

```python
training = CNN.fit(images_df, oneHotVectors_df, epochs=10, batch_size=100, callbacks=callback)
```
